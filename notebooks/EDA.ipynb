```python
# Load raw dataset
RAW_path = '../raw_analyst_ratings.csv'
RAW_df = pd.read_csv(RAW_path)
```

```python
df_ratings = RAW_df
df_ratings['date'] = pd.to_datetime(df_ratings['date'], errors='coerce')
df_ratings.head(10)
```

```python
df_ratings[df_ratings['date'].isna()]
```

```python
# Headline length cal
df_ratings['headline_length'] = df_ratings['headline'].str.len()
```

```python
# Basic statistics
df_ratings['headline_length'].describe()
```

```python
plt.figure(figsize=(10, 6))
df_ratings['headline_length'].hist(bins=20, alpha=0.7)
plt.title("Distribution of Headline Lengths (Analyst Ratings)")
plt.xlabel("Headline Length")
plt.ylabel("Frequency")
plt.show()
```

```python
df_ratings['publisher'].value_counts()
```

```python
top_publishers_ratings = df_ratings['publisher'].value_counts().head(10)
top_publishers_ratings.plot(kind='bar', figsize=(10, 6), alpha=0.7)
plt.title("Top 10 Most Active Publishers (Analyst Ratings)")
plt.xlabel("Publisher")
plt.ylabel("Number of Articles")
plt.show()
```
