### Text Analysis

```python
from textblob import TextBlob
```

```python
# Cleanup headline
df_ratings['clean_headline'] = df_ratings['headline'].str.lower()
df_ratings.head()
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
df_ratings['sentiment'] = df_ratings['clean_headline'].apply(get_sentiment)
```

```python
df_ratings.head()
```

```python
df_ratings['sentiment'].value_counts().plot(kind='bar', figsize=(10, 6), alpha=0.7, title='Sentiment Distribution')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

tfidf = TfidfVectorizer(stop_words='english', max_features=500)
tfidf_matrix = tfidf.fit_transform(df_ratings['clean_headline'])

lda = LatentDirichletAllocation(n_components=5, random_state=42)  # Adjust n_components for number of topics
lda.fit(tfidf_matrix)

# Extract topics
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

no_top_words = 10
feature_names = tfidf.get_feature_names_out()
display_topics(lda, feature_names, no_top_words)
```

## TASK 2

```python
### Apply Technical Indicators with TA-Lib
import talib

df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
# Moving Averages
df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)  # 20-day Simple Moving Average
df['EMA_20'] = talib.EMA(df['Close'], timeperiod=20)  # 20-day Exponential Moving Average

# RSI (Relative Strength Index)
df['RSI'] = talib.RSI(df['Close'], timeperiod=14)

# MACD (Moving Average Convergence Divergence)
df['MACD'], df['MACD_Signal'], df['MACD_Hist'] = talib.MACD(df['Close'],
                                                             fastperiod=12,
                                                             slowperiod=26,
                                                             signalperiod=9)
```

```python
from pynance import prices as p

# Load stock prices into PyNance
price_data = p.load(df[['Close']])

# Calculate daily returns
df['Daily_Return'] = price_data.returns()

# Calculate cumulative returns
df['Cumulative_Return'] = (1 + df['Daily_Return']).cumprod()

# Display the DataFrame with metrics
print(df[['Close', 'Daily_Return', 'Cumulative_Return']].tail())
```

```python
# Plot stock prices and moving averages
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close Price', alpha=0.7)
plt.plot(df.index, df['SMA_20'], label='20-Day SMA', alpha=0.7)
plt.plot(df.index, df['EMA_20'], label='20-Day EMA', alpha=0.7)
plt.title("Stock Prices with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()

# Plot RSI
plt.figure(figsize=(14, 5))
plt.plot(df.index, df['RSI'], label='RSI', alpha=0.7)
plt.axhline(70, color='red', linestyle='--', label='Overbought')
plt.axhline(30, color='green', linestyle='--', label='Oversold')
plt.title("Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.grid()
plt.show()

# Plot MACD
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['MACD'], label='MACD', alpha=0.7)
plt.plot(df.index, df['MACD_Signal'], label='MACD Signal', alpha=0.7)
plt.bar(df.index, df['MACD_Hist'], label='MACD Histogram', alpha=0.7)
plt.title("MACD Indicator")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()

# Plot cumulative returns
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Cumulative_Return'], label='Cumulative Return', alpha=0.7, color='purple')
plt.title("Cumulative Returns")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid()
plt.show()
```

```python
# Load datasets
news_df = pd.read_csv('raw_analyst_ratings.csv')
stock_df = pd.read_csv('AAPL_historical_data.csv')

# Parse dates
news_df['date'] = pd.to_datetime(news_df['date']).dt.date  # Convert to date only
stock_df['Date'] = pd.to_datetime(stock_df['Date']).dt.date  # Convert to date only

# Group news by date and calculate average sentiment for each day (after sentiment analysis)
news_df = news_df.groupby('date').agg({'headline': 'count', 'sentiment_score': 'mean'}).reset_index()
news_df.rename(columns={'headline': 'article_count'}, inplace=True)

# Align datasets by date
merged_df = pd.merge(stock_df, news_df, left_on='Date', right_on='date', how='inner')

# Drop redundant columns
merged_df.drop(columns=['date'], inplace=True)
```

```python
from textblob import TextBlob

# Calculate sentiment scores
def get_sentiment_score(text):
    return TextBlob(text).sentiment.polarity

news_df['sentiment_score'] = news_df['headline'].apply(get_sentiment_score)
```

```python
# Calculate daily returns
stock_df['Daily_Return'] = stock_df['Close'].pct_change() * 100

# Drop NaN values after calculating returns
stock_df.dropna(inplace=True)
```

```python
from scipy.stats import pearsonr

# Merge the datasets (already done in date alignment step)

# Calculate correlation
correlation, p_value = pearsonr(merged_df['sentiment_score'], merged_df['Daily_Return'])
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['sentiment_score'], merged_df['Daily_Return'], alpha=0.7)
plt.title('Sentiment Score vs. Daily Stock Returns')
plt.xlabel('Sentiment Score')
plt.ylabel('Daily Stock Return (%)')
plt.grid()
plt.show()
```
